```python
#Deliverable #1: Collect the Data

# Dependencies and Setup
import pandas as pd

# File to Load (Remember to change the path if needed.)
school_data_to_load = "Resources/new_full_student_data.csv"
student_data_to_load = "Resources/new_full_student_data.csv"

# File to load for School Data and Student Data and store into a  DataFrame
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Cleaning Student Names and Replacing Substrings and add prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

# Iterate through the words in the "prefixes_suffixes" list and replace them with an empty space, "".
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")

# Check names.
student_data_df.head(10)
```

```python
#Deliverable #2: Prepare the Data

#Check for any NaN values in the data, a
student_data_df.isna().sum()
```

```python
#Drop any rows that contain NaN values, and then verify that no NaN values remain, as the following code shows:
student_data_df = student_data_df.dropna()
student_data_df.isna().sum()
```

```python
#Check for any duplicated rows in the data, 
student_data_df.duplicated().sum()
```

```python
#Drop the duplicated rows, and then verify that they have been dropped
student_data_df = student_data_df.drop_duplicates()
student_data_df.duplicated().sum()
```

```python
#Check the data types of the columns (
student_data_df.dtypes
```

```python
#Note the suffixes in the grade numbers (for example, "11th" instead of "11"), and 
#use the str.replace function to remove them. Then review the column to verify the removal, as the following code shows:

student_data_df['grade'] = student_data_df['grade'].str.replace('th', '')
student_data_df['grade']
```
